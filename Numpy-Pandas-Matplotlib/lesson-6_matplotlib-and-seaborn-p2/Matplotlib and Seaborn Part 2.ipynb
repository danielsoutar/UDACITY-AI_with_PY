# 2. Scatterplots and Correlation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

fuel_econ = pd.read_csv('fuel-econ.csv')
print(fuel_econ.shape)
fuel_econ.head(6)
```

```python
plt.scatter(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', fit_reg = False)
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
fuel_econ[['displ', 'comb']].describe()
```

```python
bins_x = np.arange(0.6, 7+0.3, 0.3)
bins_y = np.arange(12, 58+3, 3)
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r', bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
import warnings

bins_x = np.arange(0.5, 10.5+1, 1)
bins_y = np.arange(-0.5, 10.5+1, 1)
h2d = plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb',
               bins = [bins_x, bins_y], cmap = 'viridis_r', cmin = 0.5)
counts = h2d[0]

# loop through the cell counts and add text annotations for each
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 7: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'black')
```

# 6. Violin Plots

```python
sedan_classes = ['Minicompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);
```

```python
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb')
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', 
             color = base_color)
plt.xticks(rotation = 15)
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', 
             color = base_color, inner = None)
plt.xticks(rotation = 15)
```

```python
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', 
             color = base_color, inner = 'quartile')
plt.xticks(rotation = 15)
```

# 9. Clustered Bar Charts

```python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x: x.split()[0])
```

```python
fuel_econ.groupby(['VClass', 'trans_type']).size()
```

```python
ct_counts = fuel_econ.groupby(['VClass', 'trans_type']).size()
ct_counts = ct_counts.reset_index(name = 'count') #converts series into a DataFrame
ct_counts = ct_counts.pivot(index = 'VClass', columns = 'trans_type', values = 'count')
```

```python
sb.heatmap(ct_counts)
```

```python
sb.heatmap(ct_counts, annot = True, fmt = 'd') # decimal values
```

```python
sb.countplot(data = fuel_econ, x = 'VClass')
plt.xticks(rotation = 15)
```

```python
sb.countplot(data = fuel_econ, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation = 15)
```

# 12. Adaptation of Univariate Plots

```python
sb.barplot(data = fuel_econ, x= 'VClass', y = 'comb', color = base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
```

```python
#errwidth = 0 removes whiskers
sb.barplot(data = fuel_econ, x= 'VClass', y = 'comb', color = base_color,
          errwidth = 0)
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
```

```python
#sd shows standard deviation
sb.barplot(data = fuel_econ, x= 'VClass', y = 'comb', color = base_color,
          ci = 'sd')
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
```

```python
#pointplot - if not interested in zero lower bound (dont need base color)
sb.pointplot(data = fuel_econ, x= 'VClass', y = 'comb', ci = 'sd')
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
```

```python
#if nominal variable set linestyle = '' (empty)
#pointplot - if not interested in zero lower bound (dont need base color)
sb.pointplot(data = fuel_econ, x= 'VClass', y = 'comb', ci = 'sd',
            linestyles = '')
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
```
