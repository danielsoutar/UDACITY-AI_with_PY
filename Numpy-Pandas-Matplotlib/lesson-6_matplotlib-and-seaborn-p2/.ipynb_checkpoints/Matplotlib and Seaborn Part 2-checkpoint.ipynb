# 2. Scatterplots and Correlation

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

fuel_econ = pd.read_csv('fuel-econ.csv')
print(fuel_econ.shape)
fuel_econ.head(6)
````

````python
plt.scatter(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
````

````python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
````

````python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', fit_reg = False)
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
````

````python
import warnings

bins_x = np.arange(0.5, 10.5+1, 1)
bins_y = np.arange(-0.5, 10.5+1, 1)
h2d = plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb',
               bins = [bins_x, bins_y], cmap = 'viridis_r', cmin = 0.5)
counts = h2d[0]

# loop through the cell counts and add text annotations for each
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 7: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.5, bins_y[j]+0.5, int(c),
                     ha = 'center', va = 'center', color = 'black')
````

# 9. Clustered Bar Charts

````python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x: x.split()[0])
````

````python
fuel_econ.groupby(['VClass', 'trans_type']).size()
````

````python
ct_counts = fuel_econ.groupby(['VClass', 'trans_type']).size()
ct_counts = ct_counts.reset_index(name = 'count') #converts series into a DataFrame
ct_counts = ct_counts.pivot(index = 'VClass', columns = 'trans_type', values = 'count')
````

````python
sb.heatmap(ct_counts)
````

````python
sb.heatmap(ct_counts, annot = True, fmt = 'd') # decimal values
````

````python
sb.countplot(data = fuel_econ, x = 'VClass')
plt.xticks(rotation = 15)
````

````python
sb.countplot(data = fuel_econ, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation = 15)
````

# 13. Lineplots

- lines emphasize relative change 
- emphasizes trends across x-values
- quantative vs quantative

````python
# errorbar plots a line between points 
# connects x and y lines without any order
plt.errorbar(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
````

````python
bins_e = np.arange(0.6, 7+0.2, 0.2) # bin edges
bins_c = bins_e[:-1] + 0.1 # bin centers so points can be plotted in accurate positions

# figures out which bin each datapoint should be used in
displ_binned = pd.cut(fuel_econ['displ'], bins_e, include_lowest = True) 

# group by displacement bins and take the mean of points in each bin
comb_mean = fuel_econ['comb'].groupby(displ_binned).mean() 

plt.errorbar(x = bins_c, y = comb_mean)
plt.xlabel('Displacement (l)')
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
````

gap in line -- use ```errorbar``` function to compute std of fuel efficiencies in each bin and plot them with ```yerr``` parameter

````python
comb_std = fuel_econ['comb'].groupby(displ_binned).std()
````

````python
plt.errorbar(x = bins_c, y = comb_mean, yerr = comb_std)
plt.xlabel('Displacement (l)')
plt.ylabel('Avg. Combined Fuel Eff. (mpg)')
````
